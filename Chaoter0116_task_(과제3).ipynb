__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
# 1. 손님이 주문한 붕어빵 종류가 재고에 있는지 확인한다.

# 2. 재고가 충분한 경우:
#    - 재고를 주문 수량만큼 차감한다.
#    - 판매 완료 메시지를 출력한다.

# 3. 재고가 부족한 경우:
#    - 재고 부족 메시지를 출력하고 재고는 차감하지 않는다.

재고 = {"팥붕어빵": 5, "슈크림붕어빵": 3, "초코붕어빵": 2}


주문_종류 = input("주문할 붕어빵 종류를 입력하세요: ")
주문_수량 = int(input("주문할 수량을 입력하세요: "))

if 주문_종류 in 재고:
    if 재고[주문_종류] >= 주문_수량:
        재고[주문_종류] -= 주문_수량
        print(f"✅ {주문_종류} {주문_수량}개 판매 완료!")
    else:  # 재고 부족
        print(f"재고 부족! 현재 {주문_종류} 재고는 {재고[주문_종류]}개 입니다.")
else:
    print(" 그런 붕어빵은 없습니다.")

print("현재 재고 상태:", 재고)
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
# 1. 무한 반복을 통해 여러 손님의 주문을 처리할 수 있도록 한다.
# 2. "종료"를 입력하면 반복을 종료한다.

# 3. 주문을 받을 때마다 [문제 2, 3]의 흐름을 따라 주문을 받고 처리한다.

# 4. 주문 처리 후 현재 재고 상태를 출력한다.

재고 = {"팥붕어빵": 5, "슈크림붕어빵": 3, "초코붕어빵": 2}

while True:
    주문_종류 = input("\n주문할 붕어빵 종류를 입력하세요 (종료 입력시 종료): ")
    if 주문_종류 == "종료":
        print("주문을 종료합니다.")
        break

    if 주문_종류 in 재고:
        주문_수량 = int(input("주문할 수량을 입력하세요: "))
        if 재고[주문_종류] >= 주문_수량:
            재고[주문_종류] -= 주문_수량
            print(f"{주문_종류} {주문_수량}개 판매 완료!")
        else:
            print(f"재고 부족! 현재 {주문_종류} 재고는 {재고[주문_종류]}개 입니다.")
    else:
        print("없는 메뉴입니다.")

    print("현재 재고 상태:", 재고)
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
# 1. 관리자 모드를 위한 반복문을 만든다.
# 2. 관리자에게 추가할 붕어빵 종류를 입력받는다.
#    - "종료" 입력 시 관리자 모드를 종료한다.

# 3. 추가할 개수를 입력받고 해당 붕어빵의 재고를 증가시킨다.

# 4. 재고 추가 후 현재 재고 상태를 출력한다.

재고 = {"팥붕어빵": 5, "슈크림붕어빵": 3, "초코붕어빵": 2}

while True :
    추가_종류 = input("관리자모드 - 추가할 붕어빵 종류를 입력해주세요.(종료를 입력할시 종료됩니다.)")
    if 추가_종류 == 0:
        print("재고추가를 종료합니다.")
        break

    추가_수량 = int(input("추가할 수량을 입력해주세요."))
    if 추가_종류 in 재고:
        재고[추가_종류] += 추가_수량
    else:
        재고[추가_종류] = 추가_수량

    print("현재 재고 상태:", 재고)
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
# 1. 각 붕어빵의 가격을 딕셔너리로 저장한다.
#    예: {"팥붕어빵": 1000, "슈크림붕어빵": 1200, "초코붕어빵": 1500}
가격표 = {"팥붕어빵": 1000, "슈크림붕어빵": 1200, "초코붕어빵": 1500}


# 2. 판매된 붕어빵의 개수를 딕셔너리로 저장한다.
#    예: {"팥붕어빵": 3, "슈크림붕어빵": 2, "초코붕어빵": 1}
판매량 = {"팥붕어빵": 3, "슈크림붕어빵": 2, "초코붕어빵": 1}

# 3. 각 붕어빵에 대해 가격 × 판매수량을 계산하여 매출을 출력한다.
총매출 = 0

for 종류, 수량 in 판매량.items():
    if 종류 in 가격표:
        매출 = 가격표[종류]*수량
        print(f'{종류}:{매출}원(수량{수량}개)')
        총매출 += 매출

print(f'총 매출 : {총매출}원')


# 4. 총 매출도 함께 계산하여 출력한다.
#    출력 예: 총 매출: 8700원
```
